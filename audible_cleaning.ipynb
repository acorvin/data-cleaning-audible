```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv(r"audible_uncleaned.csv")
df
```

```python
# Remove any duplicates from the dataset
df = df.drop_duplicates()
df
```

```python
# Remove extraneous strings in author column
df['author'] = df['author'].str.lstrip('Writtenby:')
df
```

```python
# Add a space between author first and last name
df['author'] = df['author'].str.replace(r"(?<=[a-z])(?=[A-Z])", " ",regex=True)
df
```

```python
# Remove extraneous strings in narrator column
df['narrator'] = df['narrator'].str.lstrip('Narratedby:')
df
```

```python
# Add a space between narrator first and last name
df['narrator'] = df['narrator'].str.replace(r"(?<=[a-z])(?=[A-Z])", " ",regex=True)
df
```

```python
# Extract star rating from stars column and convert it to a float
df['star_rating'] = df['stars'].str.extract(r'(\d+\.\d+|\d+)').astype('float')
df
```

```python
# Drop rows of movies that have not been rated
df = df[df['stars'] != 'Not rated yet']
df.reset_index(drop=True, inplace=True)
df
```

```python
# Drop the stars and price columns. They will not be used.
df = df.drop(columns = ['stars', 'price'])
df
```

```python
# Convert the releasedate column datetime format
df['releasedate'] = pd.to_datetime(df['releasedate'], format='%d-%m-%y')
df['releasedate'] = df['releasedate'].dt.strftime('%Y-%m-%d')
df
```

```python
# Extract hours and minutes as integers
hours_extracted = df['time'].str.extract(r'(\d+) hrs?')
minutes_extracted = df['time'].str.extract(r'(\d+) mins?')

# Convert hours and minutes to integers, filling NaN values with 0
df['hours'] = hours_extracted[0].fillna(0).astype(int)
df['minutes'] = minutes_extracted[0].fillna(0).astype(int)

# Calculate total time in minutes
df['total_minutes'] = df['hours'] * 60 + df['minutes']

# Convert total time in minutes to HH:MM:SS format
df['duration'] = df['total_minutes'].apply(lambda x: f"{x // 60:02d}:{x % 60:02d}:00")

# Drop intermediate and unused columns if needed
df.drop(['hours', 'minutes', 'total_minutes', 'time'], axis=1, inplace=True)
df
```
